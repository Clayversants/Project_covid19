```python
import pandas as pd
```

```python
#importando dados csv

#teste = 'Sales2022.csv'
caminho_arquivo = r'C:\Users\Clayver\Desktop\Fundamentos  python\Sales2022.csv'

#tranformando os tipos de dados
data_tipe = {
    'CustomerNumber' : str,
    'SalesOrder' : str
}

df = pd.read_csv(caminho_arquivo,dtype = data_tipe,parse_dates=['SalesDate'] )
```

```python
df.head(5)
```

```python
#tipos de dados
#df.dtypes
```

```python
#selecionando colunas

df.columns

colunas = ['Country', 'State', 'City','PlantName', 'MaterialNumber', 'BusinessLine', 'Category', 'SalesDate','SalesOrder', 'Volume', 'Revenue']

df[colunas].head(5)
```

```python
#filtrando dados

#df['PlantName'].unique()

sorocaba_df = df[df['PlantName'] == 'Sorocaba']
sorocaba_df.head()
```

```python
#Filtrando receita

sorocaba_df[sorocaba_df['Revenue']>=1000].head()
```

```python
#filtrando data

sorocaba_df[(sorocaba_df['SalesDate']>='2022-01-01') & (sorocaba_df['SalesDate']<='2022-01-31')].sort_values(by='SalesDate', ascending=True)
```

```python
#agrupando GROUP BY

sorocaba_df.groupby('Category')['Revenue'].sum().sort_values(ascending=False)
```

```python
sorocaba_df_group = sorocaba_df.groupby(['BusinessLine','Category'])['Revenue'].sum().sort_values(ascending=False).reset_index()
```
